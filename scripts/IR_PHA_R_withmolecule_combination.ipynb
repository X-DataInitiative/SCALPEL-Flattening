## Read the IRPHAR table

```python
import pandas as pd
```

```python
filePath = "/Users/firas/Project/SNIIRAM-flattening/src/main/resources/IR_PHA_R/IR_PHA_R.csv"
ir_pha_r = pd.read_csv(filePath, sep=";", encoding ='ISO-8859-1', na_values="NaN")

# Keep only the lines with PHA_NOM_PA
ir_pha_r = ir_pha_r[ir_pha_r.PHA_NOM_PA.notnull()].copy()

ir_pha_r.head()
```

## Add combination molecule Column 

```python
import re
data = ir_pha_r

#suppress excipents (string inside)
data["molecule_combination"] = data["PHA_NOM_PA"].apply(lambda x: re.sub(r"[\(\) ]*", "", x))

#Separate molecules  (+ ET)
data["molecule_combination"] = data.molecule_combination.apply(lambda x: re.split(" ET |\+|,", x.upper()))

#suppress leading and end spaces
data["molecule_combination"] = data.molecule_combination.apply(lambda x: [e.lstrip().strip() for e in x])

#sort molecules
data["molecule_combination"] = data.molecule_combination.apply(lambda x: sorted(x))

#join by _
data["molecule_combination"] = data.molecule_combination.apply(lambda x: "_".join(x))
```
